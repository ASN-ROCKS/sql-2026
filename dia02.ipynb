```sql
SELECT idCliente,
       qtdePontos,
       CASE
            WHEN qtdePontos = 0 THEN 'INATIVO'
            WHEN qtdePontos < 500 THEN 'ATIVIDADE BAIXA'
            WHEN qtdePontos < 5000 THEN 'ATIVIDADE MEDIA'
            ELSE 'ATIVIDADE ALTA'
       END AS flagAtividade

FROM workspace.tmw_points.clientes
```

```sql
SELECT count(*),                                    -- qtde de linhas da tabela
       count(distinct idCliente),
       count(*) / count(distinct idCliente),
       count(distinct idCliente, date(DtCriacao)),
       count(distinct date(DtCriacao))

FROM workspace.tmw_points.transacoes
```

```sql
SELECT 
       sum(CASE WHEN QtdePontos > 0 THEN QtdePontos END) AS pontosPos,
       
       count(DISTINCT CASE WHEN qtdePontos > 0 THEN idTransacao END) AS qtdtransacaoPos,
       sum(CASE WHEN qtdePontos > 0 THEN 1 else 0 END) AS qtdtransacaoPos,
       avg(CASE WHEN qtdePontos > 0 THEN 1 else 0 END) AS propTransacaoPos,
       sum(CASE WHEN qtdePontos > 0 THEN 1 END) / count(*) AS propTransacaoPos,
       sum(CASE WHEN QtdePontos < 0 THEN QtdePontos END) AS pontosNeg,
       sum(QtdePontos) as pontosSaldo

FROM workspace.tmw_points.transacoes
```
