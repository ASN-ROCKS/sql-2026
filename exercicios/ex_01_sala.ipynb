```sql
-- Selecione todos os clientes com email cadastrado

SELECT *
FROM workspace.tmw_points.clientes
WHERE flEmail = 1
```

```sql
-- Selecione todas transações de 50 pontos (exatos)
SELECT *
FROM workspace.tmw_points.transacoes
WHERE QtdePontos = 50
```

```sql
-- Selecione todos clientes com mais de 500 pontos
SELECT idCliente, qtdePontos
FROM workspace.tmw_points.clientes
WHERE qtdePontos > 500
```

```sql
-- Selecione produtos que contêm ‘churn’ no nome
SELECT *
FROM workspace.tmw_points.produtos
WHERE LOWER(DescNomeProduto) LIKE '%churn%'
```

```sql
-- 1. Lista de transações com apenas 1 ponto;
SELECT *
FROM workspace.tmw_points.transacoes
WHERE QtdePontos = 1
```

```sql
-- 2. Lista de transações realizados no fim de semana;

SELECT idTransacao,
       DtCriacao

FROM workspace.tmw_points.transacoes
-- WHERE weekday(DtCriacao) = 5 OR weekday(DtCriacao) = 6
-- WHERE weekday(DtCriacao) BETWEEN 5 AND 6
-- WHERE weekday(DtCriacao) IN (5,6)
WHERE weekday(DtCriacao) >= 5
```

```sql
-- 3. Lista de clientes com 0 (zero) pontos;
SELECT *
FROM workspace.tmw_points.clientes
WHERE qtdePontos = 0
```

```sql
-- 4. Lista de clientes com 100 a 200 pontos (inclusive ambos);
SELECT *
FROM workspace.tmw_points.clientes
-- WHERE qtdePontos >= 100 AND qtdePontos <= 200
WHERE qtdePontos BETWEEN 100 AND 200
```

```sql
-- 5. Lista de produtos com nome que começa com “Venda de”;
SELECT *
FROM workspace.tmw_points.produtos
WHERE DescNomeProduto LIKE 'Venda de%'
```

```sql
-- 6. Lista de produtos com nome que termina com “Lover”;
SELECT *
FROM workspace.tmw_points.produtos
WHERE DescNomeProduto LIKE '%Lover'
```

```sql
-- 7. Lista de produtos que são “chapéu”;
SELECT *
FROM workspace.tmw_points.produtos
WHERE LOWER(DescNomeProduto) LIKE '%chapéu%'
```

```sql
-- 9. Listar todas as transações adicionando uma coluna nova sinalizando “alto”, “médio” e “baixo”
-- para o valor dos pontos [<10 ; <500; >=500]

SELECT *,
        CASE
          WHEN qtdePontos < 10 THEN 'baixo'
          WHEN qtdePontos < 500 THEN 'médio'
          ELSE 'alto'
        END AS NivelPontos

FROM workspace.tmw_points.transacoes
```
