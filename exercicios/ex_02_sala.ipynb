```sql
-- Qual o saldo de pontos total das carteiras?

SELECT SUM(qtdePontos)
FROM workspace.tmw_points.clientes
```

```sql
-- Quantos pontos já foram subtraídos dos clientes?

SELECT sum(QtdePontos)
FROM workspace.tmw_points.transacoes
WHERE QtdePontos < 0
```

```sql
-- Qual a média de saldo em carteira?
SELECT avg(qtdePontos)
FROM workspace.tmw_points.clientes
```

```sql
-- 1. Quantos clientes tem email cadastrado?
SELECT flEmail,
       COUNT(*)

FROM workspace.tmw_points.clientes
GROUP BY flEmail
```

```sql
-- 2. Qual cliente juntou mais pontos positivos em 2025-05?

SELECT IdCliente,
       SUM(QtdePontos) AS qtdeTotalPontos
FROM workspace.tmw_points.transacoes

WHERE year(DtCriacao) = 2025 AND month(DtCriacao)=5
AND QtdePontos > 0

GROUP BY IdCliente
ORDER BY qtdeTotalPontos DESC

LIMIT 1
```

```sql
-- 3. Qual cliente fez mais transações no ano de 2024?
SELECT IdCliente,
       count(*) AS qtdeLinhas

FROM workspace.tmw_points.transacoes
WHERE year(DtCriacao) = 2024
GROUP BY IdCliente

ORDER BY 2 DESC
LIMIT 1
```

```sql
-- 4. Quantos produtos são de rpg?

select count(*)
from workspace.tmw_points.produtos
WHERE DescCategoriaProduto = 'rpg'
```

```sql
-- 5. Qual o valor médio de pontos positivos por dia?


SELECT 
       count(distinct date(DtCriacao)) AS qtdeDias,
       sum(CASE WHEN qtdePontos > 0 THEN qtdePontos ELSE 0 END) AS qtdeTotalPontos,
       sum(CASE WHEN qtdePontos > 0 THEN qtdePontos ELSE 0 END) /  count(distinct date(DtCriacao)) AS avgPontosDia

FROM workspace.tmw_points.transacoes
```

```sql
-- 6. Qual dia da semana que mais transações em 2025?
SELECT
      CASE
          WHEN weekday(DtCriacao) = 0 THEN 'Segunda Feira'
          WHEN weekday(DtCriacao) = 1 THEN 'Terça Feira'
          WHEN weekday(DtCriacao) = 2 THEN 'Quarta Feira'
          WHEN weekday(DtCriacao) = 3 THEN 'Quinta Feira'
          WHEN weekday(DtCriacao) = 4 THEN 'Sexta Feira'
          WHEN weekday(DtCriacao) = 5 THEN 'Sabado'
          WHEN weekday(DtCriacao) = 6 THEN 'Domingo'
      END AS diaSemana,
       count(*) AS qtdeTransacao

FROM workspace.tmw_points.transacoes
WHERE year(DtCriacao) = 2025

GROUP BY 1
ORDER BY qtdeTransacao DESC
```

```sql
-- 7. Qual o produto mais transacionado?

SELECT IdProduto,
       count(*) AS qtdeTransacoes
FROM workspace.tmw_points.transacao_produto

GROUP BY IdProduto

ORDER BY qtdeTransacoes DESC

LIMIT 1
```

```sql
-- 8. Qual o produto com mais pontos transacionados?

SELECT IdProduto,
       sum(CASE WHEN vlProduto < 0 THEN vlProduto * -1 ELSE vlProduto END ) AS qtdePontos

FROM workspace.tmw_points.transacao_produto

GROUP BY IdProduto

ORDER BY qtdePontos DESC
```
