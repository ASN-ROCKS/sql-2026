```sql
-- Selecione todos os vendedores cariocas e paulistas
SELECT *
FROM workspace.olist.sellers
WHERE seller_state = 'SP'                                 -- Aqui pegamos os paulistas
OR (seller_city = 'rio de janeiro' AND seller_state='RJ') -- Aqui pegamos os cariocas
```

```sql
-- Selecione produtos de perfumaria e bebes com altura maior que 5cm
SELECT *
FROM workspace.olist.products
WHERE product_category_name = 'perfumaria'
OR (product_category_name='bebes' AND product_height_cm > 5) 
```

```sql
-- 3. Lista de pedidos que ainda não foram enviados.
SELECT *
FROM workspace.olist.orders
WHERE order_status IN ('invoiced', 'processing', 'created', 'approved')
```

```sql
-- 4. Lista de pedidos que foram entregues com atraso.
SELECT *
FROM workspace.olist.orders
WHERE DATE(order_delivered_customer_date) > DATE(order_estimated_delivery_date)
```

```sql
-- 5. Lista de pedidos que foram entregues com 2 dias de antecedência.
SELECT *
FROM workspace.olist.orders
WHERE date_diff(DATE(order_estimated_delivery_date), DATE(order_delivered_customer_date)) >= 2
```

```sql
-- 6. Lista de pedidos feitos em dezembro de 2017 e entregues com atraso
SELECT *
FROM workspace.olist.orders
WHERE YEAR(order_purchase_timestamp)=2017 AND MONTH(order_purchase_timestamp)=12
AND DATE(order_delivered_customer_date) > DATE(order_estimated_delivery_date)
```

```sql
-- 7. Lista de pedidos com avaliação maior ou igual que 4
SELECT *
FROM workspace.olist.order_reviews
WHERE review_score >= 4
```

```sql
-- 8. Lista de pedidos com 2 ou mais parcelas menores que R20,00
SELECT *
FROM workspace.olist.order_payments
WHERE payment_value / payment_installments <= 20
AND payment_installments > 1
```

```sql
-- 9. Selecione todos os pedidos e marque se houve atraso ou não
SELECT *,
        CASE
          WHEN DATE(order_delivered_customer_date)>DATE(order_estimated_delivery_date) THEN 1
          ELSE 0
        END AS flAtraso

FROM workspace.olist.orders
WHERE order_status IN ('delivered', 'canceled')
```

```sql
-- 10. Selecione os pedidos/itens e defina os grupos em uma nova coluna:
-- para frete inferior à 10%: ‘10%’
-- para frete entre 10% e 25%: ‘10% a 25%’
-- para frete entre 25% e 50%: ‘25% a 50%’
-- para frete maior que 50%: ‘+50%’

SELECT *,
        CASE
          WHEN freight_value /  (freight_value + price) < 0.1 then '10%'
          WHEN freight_value /  (freight_value + price) < 0.25 then '10% a 25%'
          WHEN freight_value /  (freight_value + price) < 0.5 then '25% a 50%'
          ELSE '+50%'
        END AS grupoFrete

FROM workspace.olist.order_items
```

```sql
-- 11. Selecione a tabela olist.product :
-- Crie uma coluna nova chamada ‘descNovaCategoria’ seguindo:
-- agrupe ‘alimentos’ e ‘alimentos_bebidas’ em ‘alimentos’
-- agrupe ‘artes’ e ‘artes_e_artesanato’ em ‘artes’
-- agrupe todas categorias de construção em uma única categoria chamada ‘construção’
-- Cria uma coluna nova chamada ‘descPeso’
-- para peso menor que 2kg: ‘leve’
-- para peso entre 2kg e 5kg: ‘médio’
-- para peso entre 5kg e 10kg: ‘pesado’
-- para peso maior que 10kg: ‘muito pesado’

SELECT *,
        CASE
          WHEN product_category_name IN ('alimentos', 'alimentos_bebidas') THEN 'alimentos'
          WHEN product_category_name IN ('artes', 'artesanato') THEN 'artes'
          WHEN product_category_name LIKE '%construcao%' THEN 'construção'
          ELSE product_category_name
        END AS descNovaCategoria,

        CASE
          WHEN product_weight_g < 2000 THEN 'leve'
          WHEN product_weight_g < 5000 THEN 'médio'
          WHEN product_weight_g < 10000 THEN 'pesado'
          ELSE 'muito pesado'
        END AS descPeso

FROM workspace.olist.products
```
